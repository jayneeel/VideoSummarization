```python
import torch
from torch import nn
import torch.nn.functional as F
from torchvision import transforms, datasets
import numpy as np
import pandas as pd

# Define the transform 
train_transform = transforms.Compose([
        transforms.Resize((224,224)),             # takes PIL image as input and outputs PIL image
        transforms.ToTensor(),              # takes PIL image as input and outputs torch.tensor
        transforms.Normalize(mean=[0.4280, 0.4106, 0.3589],  # takes tensor and outputs tensor
                             std=[0.2737, 0.2631, 0.2601]),  # see next step for mean and std
    ])

valid_transform = transforms.Compose([ 
        transforms.Resize((224,224)),             
#         transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.4280, 0.4106, 0.3589],
                             std=[0.2737, 0.2631, 0.2601]), 
    ])

test_transform = transforms.Compose([
        transforms.Resize((224,224)),             
#         transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.4280, 0.4106, 0.3589],
                             std=[0.2737, 0.2631, 0.2601]), 
    ])
```

```python
from torch.utils.data import Dataset
import skimage 
from skimage import io as skio
from PIL import Image
from matplotlib import cm
import pandas as pd

class dataset(Dataset):

    def __init__(self, csv_file, root_dir, transform=None,emb_fn=None):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the frames.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        
        self.query_frame_train = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.transform = transform
        self.emb_fn = emb_fn

    def __len__(self):
        return len(self.query_frame_train) 

    

    def __getitem__(self, idx): 
        if torch.is_tensor(idx):
            idx = idx.tolist()
            
        img_name = os.path.join(self.root_dir,str(idx)+'.jpg')
                    
        # self.query_frame_train.iloc[idx, 2].split("/")[len(self.query_frame_train.iloc[idx, 2].split("/"))-2],
        # self.query_frame_train.iloc[idx, 2].split("/")[len(self.query_frame_train.iloc[idx, 2].split("/"))-1])
        img_name=img_name.replace(" ","%20")
        image = skio.imread(img_name) 
        query = self.query_frame_train.iloc[idx, 0]
        score_annotations = self.query_frame_train.iloc[idx, 3:4] 
        score_annotations = np.array(score_annotations)
        
        
        # sentence_embedding = sister.MeanEmbedding(lang="en")
        query_embed = self.emb_fn.encode(query)
        score_annotations = score_annotations.astype('float').reshape(-1, )
        # score = []
        # if score_annotations == 0:
        #   score = np.array([1.0,0.0,0.0,0.0])
        # if score_annotations == 1:
        #   score = np.array([0.0,1.0,0.0,0.0])
        # if score_annotations == 2:
        #   score = np.array([0.0,0.0,1.0,0.0])
        # if score_annotations == 3:
        #   score = np.array([0.0,0.0,0.0,1.0])
        
        sample = {'image': image, 'query': query_embed, 'score_annotations': score_annotations}
        
        if self.transform:
            sample['image'] = self.transform(Image.fromarray(sample['image']))
            sample['query'] = torch.from_numpy(sample['query'])
            sample['score_annotations'] = torch.from_numpy(sample['score_annotations'])

        # if self.transform:
        #     image = self.transform(Image.fromarray(image))
        #     query = torch.from_numpy(query_embed)
        #     score_annotations = torch.from_numpy(score_annotations)

        X = sample['image'],sample['query']
        y = sample['score_annotations'].type(torch.LongTensor)

        #return list(sample.values())
        return X,y
```

```python
class QVSmodel(nn.Module):
  
  def __init__(self,feature_extractor,y_shape):
    super(QVSmodel, self).__init__()
    self.feature_extractor  = feature_extractor
    x_shape = [*(feature_extractor(torch.randn(1,3,224,224))).shape][1]
    self.fc_text = torch.nn.Linear(y_shape,x_shape)
    self.fc1 = torch.nn.Linear(x_shape, 4)
    
  def forward(self, x):
    x1 = self.feature_extractor(x[0])
    y1 = F.relu(self.fc_text(x[1]))
    t1 = torch.mul(x1, y1)
    t1 = self.fc1(t1)
    return t1
```
